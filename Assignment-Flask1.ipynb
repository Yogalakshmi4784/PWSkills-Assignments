#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

- Flask's advantages include its minimalistic nature, which allows developers to choose and integrate components as needed, promoting simplicity and maintainability. 
- It also has a large and active community, extensive documentation, and a wealth of extensions, making it a powerful tool for rapid web development while giving developers the freedom to customize their projects according to their specific requirements.

#### Q3. What is App routing in Flask? Why do we use app routes?

##### In Flask, app routing refers to the process of defining URL routes for your web application. These routes specify the URLs at which your application will respond to user requests and define the corresponding functions to execute when a specific URL is accessed. 

##### App routes are essential for directing incoming requests to the appropriate view functions, enabling the application to display different content or perform various actions based on the URL requested. This routing mechanism is fundamental for building organized and functional web applications in Flask.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

##### In Flask, the url_for() function is used for URL building. It generates URLs for a given view function by referencing its name and any arguments it requires. This is especially useful when you want to avoid hardcoding URLs in your application, making it more maintainable and flexible.

##### Here's a Python code example demonstrating how to use url_for():

from flask import Flask,url_for

from flask import request

app = Flask(__name__)

@app.route("/")

def company():
    
    Company_Name= 'ABC Corporation'
    
    Location= 'India'
    
    Contact_Detail= '999-999-9999'
    
    return "Company Name: {} <br> Location: {} <br> Contact Detail: {}".format(Company_Name,Location,Contact_Detail)

@app.route("/welcome/<company_name>")

def welcome(company_name):
    
    return "Welcome to {}".format(company_name)



if __name__ == '__main__':
    
    with app.test_request_context():
        
        # Generating URLs using url_for
        
        print(url_for('company'))  # Output: '/'
        
        print(url_for('welcome',company_name='ABCCorporation'))  # Output: '/welcome/ABCCorporation'

if __name__=="__main__":
    
    app.run(host="0.0.0.0")
