#### Q1. Explain GET and POST methods.

##### - GET method is used to pass input through URL
##### - POST method is used to pass input through forms in a secure manner.

#### Q2. Why is request used in Flask?

##### The request object in Flask is used to access and manipulate data from incoming HTTP requests made to a Flask web application. It allows developers to retrieve form data, query parameters, headers, and other information sent by clients. This data is essential for processing user input, making decisions, and customizing responses in the web application.

#### Q3. Why is redirect() used in Flask?

##### 'redirect()' is used in Flask to perform a redirection to a different URL within a web application. This function is essential for handling user interactions and guiding them to the appropriate pages or routes. Redirects are commonly used after processing a form submission, ensuring that users are directed to a success page, login page, or any other relevant location, improving the user experience and flow of the application.

#### Q4. What are templates in Flask? Why is the render_template() function used?

##### Templates in Flask are files that contain HTML markup with placeholders for dynamic content. They allow developers to separate the presentation (HTML structure) from the application logic. The render_template() function in Flask is used to render these templates, dynamically filling in placeholders with data and generating complete HTML pages to be sent as responses to clients. This separation of concerns improves code maintainability and reusability while enabling the creation of dynamic and data-driven web pages in Flask applications.
