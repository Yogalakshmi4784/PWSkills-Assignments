#### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

##### Some common functions with which we can manipulate data are: iloc,sort_values, head, tail, dropna,fillna,isnull. An example for each of this is given below:

```python
players=pd.read_csv('Players.csv')
```

```python
players.head()
```

```python
players.tail()
```

```python
players.iloc[0:2,5:8] # First 2 rows of columns 5 to 7
```

```python
players.sort_values('Player')
```

```python
players.isnull().sum()  # Displays null values for each column
```

```python
players1=players.dropna()
```

```python
players1.isnull().sum() ## Rows having null values are dropped
```

#### Q7. Create a DataFrame using multiple Series. Explain with an example.

##### DataFrame can be created using multiple Series by passing those Series objects into the DataFrame constructor. Each Series will become a column in the resulting DataFrame.

```python
series1=pd.Series(['Alice','Bob','Claire'],name='NAme')
series2=pd.Series([25,30,27],name='Age')
series3=pd.Series(['Female','Male','Female'],name='Gender')
```

```python
df1=pd.DataFrame([series1,series2,series3])
df1=df1.transpose()
```
